# Introduction to Python and Jupyter Notebook

## Getting Started: Basic Steps

In a Jupyter Notebook, work is organized into **cells**. This allows you to write code in segments and see the results immediately.

### 1. Cell Types

There are two main types of cells used in Jupyter Notebook:

* **Code cells**: This is where you write Python code. When you press *Shift + Enter*, the code is executed.

* **Markdown cells**: These are used for explanations, headings, and lists.

### 2. Essential Keyboard Shortcuts

To work more efficiently, it is recommended to learn these shortcuts:

Shortcut | Action
    
**Shift + Enter**   Run the current cell and move to the next one.
    
**Alt + Enter**  Run the cell and insert a new empty cell below.
    
**Esc** then **A**  Insert a new cell **Above** the current one.
    
**Esc** then **B**  Insert a new cell **Below** the current one.
    
**Esc** then **M**  Change cell type to **Markdown**.
    
**Esc** then **Y**  Change cell type to **Code**.

# Introduction to Python for Engineering Applications

This notebook covers the basics of Python programming, NumPy for numerical calculations, and Matplotlib for data visualization. These tools are essential for analyzing harmonic oscillations and mechanical systems.

## 1. Variables and Mathematical Operations
In engineering, we start by defining physical parameters such as mass and stiffness.

```python
import numpy as np

# System Parameters
m = 10.5        # Mass in kg
k = 500.0       # Stiffness in N/m

# Calculate Natural Frequency (omega_n = sqrt(k/m))
omega_n = np.sqrt(k / m)
print(f"Natural Frequency: {omega_n:.2f} rad/s")
```

## 2. Arrays with NumPy
NumPy allows us to create vectors of time and calculate displacement efficiently.

```python
import numpy as np

# Create a time vector from 0 to 5 seconds with 1000 points
t = np.linspace(0, 5, 1000)

# Define displacement x(t) = A * cos(omega_n * t)
A = 0.05  # Amplitude in meters
x = A * np.cos(omega_n * t)

print("Time and displacement arrays created successfully.")
```

## 3. Visualization with Matplotlib
Visualizing the wave helps us understand the oscillation period and amplitude.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(t, x, label='Free Undamped Vibration', color='blue')
plt.title('Displacement vs Time')
plt.xlabel('Time [s]')
plt.ylabel('Displacement [m]')
plt.grid(True)
plt.legend()
plt.show()
```
