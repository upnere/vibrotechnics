How do we interpret $\omega_1$ and $\omega_2$ from this graph?

**Low RPM ($\omega_1 = 60$ RPM)**: This point is close to the resonance zone (so the amplitude is higher, about $7.35$ mm). Here, the system's response is dictated by stiffness and damping. Small changes in RPM cause large changes in vibration.

**High RPM ($\omega_2 = 1000$ RPM)**: Here, the system has "passed" resonance. The amplitude tends to the theoretical limit or asymptote ($m_0 e / m$). In this zone, the vibration is no longer affected by springs or dampers, but only by the mass ratio. This is a safer operating mode for stable operation.
